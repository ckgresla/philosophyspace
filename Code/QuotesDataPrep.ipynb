# Quotes Data Prep

Following EDA & BertServ testing, clean Quotes & Get Embeddings for all Quotes

```python
import pandas as pd 

# Read in & LowerCase all strings (using lowercase BERT)
df = pd.read_csv("../Data/COMBINED-QUOTES.csv", engine="python")
df["author"] = df["author"].str.lower() #Team Added Author Strings not LowerCase
df["quote"] = df["quote"].str.lower()

df
```

```python
# Get Embeddings per Quote
from bert_serving.client import BertClient
bc = BertClient()

# Vars
embed_list = []
n_remaining = df["quote"].nunique()


for quote in df["quote"]:
    # Print Info per 100 embeddings
    n_remaining -= 1
    if n_remaining % 100 == 0:
        print(n_remaining, "Quotes Remaining") #need python 3.6 or newer for f-strings
    
    # Send to BertServ
    embedding = bc.encode([quote])
    embed_list.append(embedding)

df["embedding"] = embed_list


# Write Out
df.to_csv("../Data/Quotes+Embeddings.csv", index=False)
```
