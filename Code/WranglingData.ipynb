```python
# Check for Authors with High Similarity 
import textdistance
hmap = {}


for a1 in dfc["author"].unique():
    for a2 in dfc["author"].unique():
        # Make sure not exact same author
        if a1 != a2:
            try:
                if hmap[a2] == a1:
                    continue
                elif hmap[a1] == a2:
                    continue
                else:
                    similarity = textdistance.jaro_winkler(a1, a2)
                    if similarity > 0.95:
                        a1l = len(a1)
                        a2l = len(a2)
                        if a1l < a2l:
                            hamp[a1] = a2 #save pair for usage later
                        else: 
                            hamp[a2] = a1 #save pair for usage later
                        print(f"{a1} : {a2}    {similarity:.2f}")
            except:
                pass

```

Made use of [Jaro-Winkler](https://en.wikipedia.org/wiki/Jaro%E2%80%93Winkler_distance) distance in above (a string similarity metric, is not technically a distance metric but is a measure of similarity between two strings) -- this is a bit sneaky because without checking the author name, quotes would be attributed to different people! The set of quotes we ran this on was atomic (dropped all dupes in lower form irregardless of who wrote the quote)
